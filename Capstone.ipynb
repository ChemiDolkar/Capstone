```python
df=pd.DataFrame(np.random.rand(50,2),columns=['y1_grads_earn','label_degree_level'])
df.plot.scatter(x='label_degree_level',y='y1_grads_earn');
```

This chart shows correlation between earnings after one year of graduation and degree level. It’s clustered all over the chart pretty much representing no positive correlation between the two variables. From my point of view, it could be mainly because of the two reasons. 1) It shows degree level not the field of study they are graduated with. For example, a graduate of bachelor’s degree in Engineering might be earning more than a graduate of master’s degree in International Development. 2) There are lot of missing values in the data that could have some impact on the outcome.

```python
df.groupby(['Degree'])['y1_grads_earn'].count().sort_values(ascending=False)
```

```python
df.groupby(['Degree'])['y1_grads_earn'].count().sort_values(ascending=False).plot(kind="bar")
```

This chart shows graduates of bachelor’s and associate degrees earning more than other degrees. This could be also because of the same two reasons from the previous chart. And in addition to that the 3rd reason might be because there are a greater number of graduates with bachelor’s and associate degrees than other degrees.
